```python
import pandas as pd

# Cargamos el archivo
df = pd.read_csv('positions.csv')

columns = ['time(s)', 'vehicle_id','collision']
df = df[columns]
df = df.sort_values(by=['vehicle_id', 'time(s)','collision'])
df.head()
```

```python
# Separamos en colisiones y no colisiones
df_col = df.loc[df['collision'] == 1,]
df_nocol = df.loc[df['collision'] == 0,]
```

```python
print(df_col.shape)
print(df_nocol.shape)
```

```python
import os
import pickle

## Cargamos vehicles_id
VEHICLE_IDS_FILE = os.path.join('./', "vehicle_ids.pkl")
with open(VEHICLE_IDS_FILE, 'rb') as f_vehicle_ids:
    vehicle_ids = pickle.load(f_vehicle_ids)
    
# Nos quedamos únicamente con los vehículos que esten en vehicles_id
df_col = df_col[df_col['vehicle_id'].isin(vehicle_ids)]
df_nocol = df_nocol[df_nocol['vehicle_id'].isin(vehicle_ids)]

# Como este proceso suele tardar, los guardamos como csv para próximas veces ahorrarnos el tiempo
df_col.to_csv('df_col.csv')
df_nocol.to_csv('df_nocol.csv')
```

```python
print(df_col.shape)
print(df_nocol.shape)
```

```python
# Crear dataset de no colisiones: vehículos que coinciden en el tiempo final y colisionan
import random

cont = 0
l_time = list(df_nocol['time(s)'].unique())

while cont < 5000:
    l = []
    print(cont)
    while len(l) < 2:
        # Cogemos un tiempo cualquiera
        x = round(random.sample(l_time,1)[0],0)
        
        # Buscamos los vehículos en él
        l = list(df_nocol.loc[(df_nocol['time(s)'] == x),'vehicle_id'].unique())
        
        # Eliminamos este tiempo
        l_time.remove(x)
        
    # Para todas las parejas, comprobamos si hay o no intersección
    
    # Sacamos todas las posibles parejas
    res = [(a, b) for idx, a in enumerate(l) for b in l[idx + 1:]]
    
    # Para cada pareja..
    for i in range(len(res)):
        v1_id, v2_id = res[i][0], res[i][1]

        v1_ts_set = set(df_nocol[df_nocol['vehicle_id'] == v1_id].iloc[-5:, 0])
        v2_ts_set = set(df_nocol[df_nocol['vehicle_id'] == v2_id].iloc[-5:, 0])
        intersection = sorted(list(v1_ts_set.intersection(v2_ts_set)))
        if len(intersection) >= 2 :
            print(v1_id,',', v2_id)
            cont += 1
            
```
