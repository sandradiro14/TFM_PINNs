```python
import pandas as pd

# Cargamos el archivo
df = pd.read_csv('positions.csv')

columns = ['time(s)', 'vehicle_id', 'victim_id','collision']
df = df[columns]
df.head()
```

```python
# Cogemos las colisiones
df_col = df.loc[df['collision'] == 1,]

# Cogemos las que ambos estén en vehículos_id
import os
import pickle

# ## Cargamos vehicles_id
# VEHICLE_IDS_FILE = os.path.join('./', "vehicle_ids.pkl")
# with open(VEHICLE_IDS_FILE, 'rb') as f_vehicle_ids:
#     vehicle_ids = pickle.load(f_vehicle_ids)
    
# Generamos vechicles_id con todos los que hay
vehicle_ids = []
distinct_vehicle_id = set(df.loc[:,'vehicle_id'])
distinct_victim_id = set(df.loc[:,'victim_id'])
set_vehicles = distinct_vehicle_id | distinct_victim_id
vehicle_ids = list(set_vehicles)
    
# Nos quedamos únicamente con los vehículos que esten en vehicles_id
df_col = df_col[df_col['vehicle_id'].isin(vehicle_ids)]
df_col = df_col[df_col['victim_id'].isin(vehicle_ids)]
```

```python
print(len(vehicle_ids))
print(df_col.shape)
```

```python
# Guardamos el nuevo vehicle_ids
VEHICLE_IDS_FILE = os.path.join('./', "vehicle_ids.pkl")
with open(VEHICLE_IDS_FILE, "wb") as f_vehicle_ids:
    pickle.dump(vehicle_ids, f_vehicle_ids)
```
